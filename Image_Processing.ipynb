#C. Image Compression

##Method 1: Lossy compression method

#Load the original image for display

```python
# step 1: Load the original image to be processed
image_path = '/content/drive/My Drive/classroom_building.jpg'
classroom_building_image = Image.open(image_path)

# step 2:  Display the original image
plt.imshow(classroom_building_image)
plt.axis('off')
plt.title("College of Informatics building Original Image")
plt.show()

# step 3: Get file size in bytes
original_size_bytes = os.path.getsize(image_path)

# step 4: Convert the size to KB and MB
original_size_kb = original_size_bytes / 1024
original_size_mb = original_size_kb / 1024

# step 5: Print the original image sizes
print(f"Original Image Size: {original_size_bytes} bytes")
print(f"Original Image Size: {original_size_kb:.2f} KB")
print(f"Original Image Size: {original_size_mb:.2f} MB")
```

#Lossy compression

```python
# Save the image with lossy compression
output_path = '/content/drive/My Drive/lossy_classroom_building.jpg'
classroom_building_image.save(output_path, 'JPEG', quality=50)

print(f"Image saved with lossy compression to {output_path}")
```

#Load the compressed image for display

```python
# step 1: Load the lossy compressed image
lossy_classroom_building = Image.open(output_path)

# step 2: Display the lossy compressed image
plt.imshow(lossy_classroom_building)
plt.axis('off')
plt.title("Image  of informatics classrooms after compression")
plt.show()

# step 3: Get file size in bytes
lossy_classroom_building_sizes = os.path.getsize(output_path)

# step 4: Convert the size to KB and MB
lossy_size_kb = lossy_classroom_building_sizes / 1024
lossy_size_mb = lossy_size_kb / 1024

# step 5: Print the lossy compressed image sizes
print(f"Lossy Compressed Image Size: {lossy_classroom_building_sizes} bytes")
print(f"Lossy Compressed Image Size: {lossy_size_kb:.2f} KB")
print(f"Lossy Compressed Image Size: {lossy_size_mb:.2f} MB")
```

#Size reduction rate after lossy compression

```python
#compute the size reduction of the image in percentage
# Calculate the reduction in size
size_reduction_bytes = original_size_bytes - lossy_classroom_building_sizes
size_reduction_percentage = (size_reduction_bytes / original_size_bytes) * 100

# Print the comparison results
print("Size reduction after lossy compression:\n")
print(f"Size Reduction in Bytes: {size_reduction_bytes} bytes")
print(f"Size Reduction in percentage(%): {size_reduction_percentage:.2f}%")
```

#D. Morphological image processing

```python
# step 1: Load the image
image = cv2.imread('/content/drive/My Drive/lib_front.jpg')

# step 2: Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# step 3: Apply binary thresholding
_, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)

# step 4:Define a kernel
kernel = np.ones((5, 5), np.uint8)

```

#Erosion

```python
erosion = cv2.erode(binary, kernel, iterations=1)
```

#Opening


```python
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
```

#Closing

#Dilation

```python
dilation = cv2.dilate(binary, kernel, iterations=1)
```

#Displaying the results after morphological image processing

```python
#Display the results
plt.figure(figsize=(12, 10))


plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(binary, cmap='gray')
plt.title('Binary Image')
plt.axis('off')


plt.subplot(2,2,3)
plt.imshow(erosion, cmap='gray')
plt.title('Eroded Image')
plt.axis('off')


plt.subplot(2,2,4)
plt.imshow(dilation, cmap='gray')
plt.title('Dilated Image')
plt.axis('off')

plt.show()
```
